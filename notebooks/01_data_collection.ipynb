```python
import yfinance as yf
```

```python
# Define the stock ticker and time period
ticker = 'AAPL'   # Apple Inc.
start_date = '2018-01-01'   # Start date for the data
end_date = '2023-01-01'     # End date for the data
```

```python
# Download stock data
stock_data = yf.download(tickers=ticker, start=start_date, end=end_date)
```

```python
len(stock_data)
```

```python
stock_data.head(20)
```

```python
# Save the stock data to a csv file
stock_data.to_csv(f'../data/raw/{ticker}_stock_data.csv')
```

## Data Collection Summary

In this phase, a historical stock data was collected from Yahoo Finance using the `yfinance` library. Specifically, 1,259 daily data points were obtained by downloading Apple Inc. (AAPL) stock prices between January 1, 2018, and January 1, 2023.

The data includes the following features:
- **Date:** The trading date
- **Open:** The price at market open
- **High:** The highest price during the trading day
- **Low:** The lowest price during the trading day
- **Close:** The price at market close
- **Volume:** The number of shares traded
- **Adj Close:** Adjusted closing price for splits/dividends

The dataset was saved as a `csv` file.
